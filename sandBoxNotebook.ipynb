```python
import numpy as np
import itertools
```

```python
# calculate the profit made from choosing some suitcase
def profit_made(multiplier, inhabitants, global_dist, briefcase_price):

    ''' Parameters
    
        multiplier : int ; multipler that is applied to every case. for e.g. x80

        inhabitants : int ; num of players choosing a breifcase
        
        global_dist : int ; the distribution we get for a specific briefcase. For e.g
                            briefcase A1 is 7.5% likely to be opened. NOTE this is a percentage
                            we need to divide by a 100.

        briefcase_price : int ; estimated price of briefcase #2 and #3 we decide. This is always
                                zero (0) for briefcase #1 since it is free to open.
    '''
    raw_val = (10000 * multiplier) / (inhabitants * (global_dist / 100))

    expected_profit = raw_val - briefcase_price

    return expected_profit

#gets all the combination of briefcases choose 3 
def get_combinations(input_list):
  '''Parameter
  
    input_list: list ; The list of all suitcase probabilites choose 3
  '''
  return list(itertools.combinations(input_list, 3))
```

```python
# cost of case 2 and 3. ***SUBJECTED TO CHANGE***
case2_cost = 2000
case3_cost = 3000

# got these from the imc website. DONT CHANGE.
multiplier = {
    'A1': 80,
    'A2': 50,
    'A3': 83,
    'A4': 31,
    'A5': 60,
    'B1': 89,
    'B2': 10,
    'B3': 37,
    'B4': 70,
    'B5': 90,
    'C1': 17,
    'C2': 40,
    'C3': 73,
    'C4': 100,
    'C5': 20,
    'D1': 41,
    'D2': 79,
    'D3': 23,
    'D4': 47,
    'D5': 30
}

# distributions of each briefcase. Got these from gpt to have
# atleast some sensible pattern to it. ***SUBJECTED TO CHANGE***
global_opening_percent = {
    'A1': 12.5,
    'A2': 5.3,
    'A3': 7.8,
    'A4': 10.2,
    'A5': 15.0,
    'B1': 4.8,
    'B2': 6.7,
    'B3': 8.5,
    'B4': 11.1,
    'B5': 13.0,
    'C1': 9.3,
    'C2': 4.0,
    'C3': 6.2,
    'C4': 7.0,
    'C5': 14.5,
    'D1': 3.8,
    'D2': 5.5,
    'D3': 7.2,
    'D4': 6.8,
    'D5': 4.1,
    'E1': 10.9,
    'E2': 12.3,
    'E3': 6.5,
    'E4': 8.0,
    'E5': 9.7
}

# Number of contestants on each briefcase. ***SUBJECTED TO CHANGE***
contestants = {
    'A1': 6,
    'A2': 4,
    'A3': 7,
    'A4': 2,
    'A5': 4,
    'B1': 8,
    'B2': 1,
    'B3': 3,
    'B4': 4,
    'B5': 10,
    'C1': 1,
    'C2': 3,
    'C3': 4,
    'C4': 15,
    'C5': 2,
    'D1': 3,
    'D2': 5,
    'D3': 2,
    'D4': 3,
    'D5': 2
}
```

```python
# get the coordinates of each briefcase in a list4. For e.g. A1, A2, A3, ...
suitcase_names = list(multiplier.keys())

# get all the possible combinations
combinations = get_combinations(suitcase_names)

# sanity check
print([combinations[i] for i in range (3)]) # cout: ('A1', 'A2', 'A3'), ('A1', 'A2', 'A4'), ('A1', 'A2', 'A5')
print(len(combinations)) # cout: 1140
```

```python
best_case_comb = []
max_profit = float('-inf')  # set to -ve infitnity to handle all possible profit values safely
for comb in combinations:
    # initialize/reset variable
    temp_profit = 0
    
    ''' NOTE:
        1) 'comb' is a list with len 3. So I harded coded index values (0, 1, 2).
        2) 'profit_made' method has these parameters: multiplier, inhabitants, global_dist, briefcase_price
        For further explantion of these paramerters look at the method declaration. 
    '''

    # briefcase #1
    case_name = comb[0]
    temp_profit += profit_made(multiplier[case_name], contestants[case_name], global_opening_percent[case_name], 0) # briefcase_price = 0 since this is case 1

    # briefcase #2
    case_name = comb[1]
    temp_profit += profit_made(multiplier[case_name], contestants[case_name], global_opening_percent[case_name], case2_cost)

    # briefcase #3
    case_name = comb[2]
    temp_profit += profit_made(multiplier[case_name], contestants[case_name], global_opening_percent[case_name], case3_cost)

    # if the this combination yields the best results then make that the max_profit.
    # Also, update the best case combination that gave that result
    if temp_profit > max_profit:
        max_profit = temp_profit
        best_case_comb = comb

print(f"The max profit is {max_profit:.4f} from the combination {best_case_comb}")
```
